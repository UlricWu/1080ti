```python
import requests
from bs4 import BeautifulSoup

class DGP:
    def __init__(self,urls):
        self.urls = urls
    def crawler(self):
        req = requests.get(self.urls)
        soup = BeautifulSoup(req.text,'lxml')
        web=[]
        temp = []
        #would return all the website of artists
        for link in soup.find_all('a'):
            if not link.get('href'): continue
            if link.get('href')[:4]=='http':
                temp = link.get('href')
            else:
                temp = self.urls+link.get('href')
            if temp.find("view") >0: web.append(temp)
        # del the last item that is not needed. https://www.straussart.co.za/artists/live/a-b/review-2018
        del web[-1]
        #del the repeated links
        self.web = list(set(web))
    def name(self):
        date = []
        artist = []
        #for each link, return the name and the BOD of the artist
        for item in self.web:
            req = requests.get(item)
            soup = BeautifulSoup(req.text,'lxml')
            artist.append(soup.h2.contents[0])
            #Replace the no information with Na
            try:
                date.append(soup.p.contents[0])
            except:
                date.append('Na')
        self.artist = artist
        self.date = date
```

```python
# the website sorts 7 categories by the alphabet
alpha = ['a-b','c-f','g-k','l-m','n-r','s-t','u-z']
url = 'https://www.straussart.co.za/artists/live/'
final_artist = []
final_date = []
temp = []
for item in alpha:
    temp = DGP(url+item)
    temp.crawler()
    temp.name()
    final_artist.append(temp.artist)
    final_date.append(temp.date)
```

```python
#the final_artist is a list in a list. it needs to be transformed into a list. So is the final_date
artist = [y for x in final_artist for y in x]
date = [y for x in final_date for y in x]
```

```python
#all I need the artists is from South African, rather than other states.
import pandas as pd
df = {'Name':artist,'Date':date}
df = pd.DataFrame(df)
final = df[df['Date'].str.match('South African')]
BOD  = final['Date'].str.replace('South African','')
final = final.assign(BOD =BOD)
final = final.drop(columns ='Date')
final
```

```python
# save as csv
final.to_csv('final_artist_bod.csv',index = False)
```
