## hailstone is the one that 
## https://en.wikipedia.org/wiki/Collatz_conjecture
## geneneral hailstone is the one that a and b is unlimited from the range 0 to 10
## \begin{equation}
f(x)=
\begin{cases}
x/2& \text{x=even}\\
a*x+b& \text{x=odd}
\end{cases}
\end{equation}

## try to find how many a and b pair to make x cyclable



```python
def hailstone(item, a, b ):

    if item % 2 == 0:
        return item//2
    if item % 2 == 1:
        return a*item+b

def numbers(a, b):
    lst = [1]
    seen = set()
    seen.add(lst[0])
    count = 0
    
    
    for _ in range(1,1000):
        temp = hailstone(lst[-1], a, b)
        temp_len_seen = len(seen) 
        seen.add(temp)
        lst.append(temp)
        count += 1
        ## identify if the set of 99th and 100th are the same.
        while count % 100 == 0:
            if len(seen) == temp_len_seen:
                return "cycle"
            else: break

    return "incycle"

def generalized():
    results = {}
    for a in range(0,11):
        for b in range(0,11):
            results[(a,b)] = numbers(a, b)
    return results
results = generalized()
results
            
```

```python
counts = {'cycle':0, 'incycle':0}
for _, value in results.items():
    counts[value] += 1
counts
```

## try more general

\begin{equation}
f(x)=
\begin{cases}
x/3& \text{x%3=0}\\
a*x+b& \text{x%3=1}\\
c*x+d& \text{x%3=2}
\end{cases}
\end{equation}

## ## try to find how many  pairs to make x cyclable

```python
def hailstone(item, a, b, c, d ):

    if item % 3 == 0:
        return item//3
    if item % 3 == 1:
        return a*item+b
    if item % 3 == 2:
        return c*item+b

def numbers(a, b, c, d):
    lst = [1]
    seen = set()
    seen.add(lst[0])
    count = 0
    
    
    for _ in range(1,1000):
        temp = hailstone(lst[-1], a, b, c, d)
        temp_len_seen = len(seen) 
        seen.add(temp)
        lst.append(temp)
        count += 1
        
        while count % 100 == 0:
            if len(seen) == temp_len_seen:
                return "cyclable"
            else: break
    return "incyclable"

def generalized():
    results = {}
    for a in range(0,11):
        for b in range(0,11):
            for c in range(0, 11):
                for d in range(0,11):
                    results[(a,b,c,d)] = numbers(a, b,c,d)
    return results
results = generalized()
results
            
```

```python
counts = {'cyclable':0, 'incyclable':0}
for _, value in results.items():
    counts[value] += 1
counts
```
